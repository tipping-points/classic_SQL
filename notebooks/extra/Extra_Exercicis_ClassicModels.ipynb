4. Calcula el promedio facturado por mes durante el 2003.

6. Lista a los clientes con más ventas (cantidad de productos por precio) en cada país.

```python
q="""
WITH aux AS (
SELECT count(productCode),productline,productVendor,
 RANK() OVER(PARTITION BY productline ORDER BY COUNT(productCode) desc) rnk
FROM products
GROUP BY productVendor,productline)
SELECT * FROM aux

"""
pd.read_sql(q,con)
```

```python
q="""
SELECT count(productCode),productline,productVendor
FROM products
GROUP BY productVendor,productline
ORDER BY count(productCode) DESC
"""
pd.read_sql(q,con)
```
